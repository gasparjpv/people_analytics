```python
#Cor ou raça
V2010_map = {
    'Branca':'Branca',
    'Preta'	:'Preta/Parda',
    'Amarela':'Outros',
    'Parda'	:'Preta/Parda',
    'Indígena':'Outros',
    'Ignorado':'Outros'
}

df['Raça'] = df['Raça'].map(V2010_map)

print(df['Raça'].unique())
```

```python
count_identification = df['Identificação por domicilio'].count()
print('Pessoa responsável: ', count_identification)

count_homem = ((df['Sexo'] == 'Homem') & (df['Estado'] == 'Rondônia')).sum()
print(f"O número total de linhas onde 'Sexo' é 'Homem' é: {count_homem}")

count_mulher = ((df['Sexo'] == 'Mulher')& (df['Estado'] == 'Rondônia')).sum()
print(f"O número total de linhas onde 'Sexo' é 'Mulher' é: {count_mulher}")


count_mulher_raca = ((df['Sexo'] == 'Mulher') & (df['Raça'] == 'Branca') & (df['Estado'] == 'Rondônia')).sum()
print(f"O número total de linhas onde 'Sexo' é 'Mulher' e 'Raça' é 'Branca' é: {count_mulher_raca}")

count_mulher_raca1 = ((df['Sexo'] == 'Mulher') & (df['Raça'] == 'Preta/Parda') & (df['Estado'] == 'Rondônia')).sum()
print(f"O número total de linhas onde 'Sexo' é 'Mulher' e 'Raça' é 'Preta/Parda' é: {count_mulher_raca1}")

count_mulher_raca2 = ((df['Sexo'] == 'Mulher')& (df['Raça'] == 'Outros')& (df['Estado'] == 'Rondônia')).sum()
print(f"O número total de linhas onde 'Sexo' é 'Mulher' e 'Raça' é 'Outros' é: {count_mulher_raca2}")

count_homem_raca = ((df['Sexo'] == 'Homem') & (df['Raça'] == 'Branca') & (df['Estado'] == 'Rondônia')).sum()
print(f"O número total de linhas onde 'Sexo' é 'Homem' e 'Raça' é 'Branca' é: {count_mulher_raca}")

count_homem_raca1 = ((df['Sexo'] == 'Homem') & (df['Raça'] == 'Preta/Parda') & (df['Estado'] == 'Rondônia')).sum()
print(f"O número total de linhas onde 'Sexo' é 'Homem' e 'Raça' é 'Preta/Parda' é: {count_mulher_raca1}")

count_homem_raca2 = ((df['Sexo'] == 'Homem')& (df['Raça'] == 'Outros')& (df['Estado'] == 'Rondônia')).sum()
print(f"O número total de linhas onde 'Sexo' é 'Homem' e 'Raça' é 'Outros' é: {count_mulher_raca2}")

total = count_homem+count_mulher
print("Total Homem + Mulher ", total)

total2 = count_mulher_raca +count_mulher_raca1 + count_mulher_raca2 + count_homem_raca + count_homem_raca1 + count_homem_raca2
print("Total Raça  ", total2)
```
